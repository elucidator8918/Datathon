```python
df_close = hdfc['Close']
df_close.plot(kind='kde')
```

```python
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
test_stationarity(df_close)
```

```python
result = seasonal_decompose((hdfc['Close'][hdfc['Close']!=0]).to_numpy(), model='multiplicative',period=30)
fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(16, 9)
```

```python
df_log = np.log(df_close)
moving_avg = df_log.rolling(12).mean()
std_dev = df_log.rolling(12).std()
plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(std_dev, color ="black", label = "Standard Deviation")
plt.plot(moving_avg, color="red", label = "Mean")
plt.legend()
plt.show()
```

```python
train_data, test_data = df_log[3:int(len(df_log)*0.9)], df_log[int(len(df_log)*0.9):]
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Dates')
plt.ylabel('Closing Prices')
plt.plot(df_log, 'green', label='Train data')
plt.plot(test_data, 'blue', label='Test data')
plt.legend()
```

```python
model_autoARIMA = auto_arima(train_data, start_p=0, start_q=0,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=10, max_q=10, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=True,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
print(model_autoARIMA.summary())
model_autoARIMA.plot_diagnostics(figsize=(15,8))
plt.show()
```

```python
model = smapi.tsa.arima.ARIMA(train_data, order=(4,0,10))
result = model.fit()
print(result.summary())
```

```python
res=result.forecast(25,alpha=0.01).to_frame()
test_data=test_data.to_frame()
```

```python
from sklearn.metrics import mean_squared_log_error
RMSLE =  np.sqrt(mean_squared_log_error(test_data['Close'].to_numpy(),res['predicted_mean'].to_numpy()))
print('RMSLE:',RMSLE)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
sales = hdfc['Close']

model = ARIMA(sales, order=(4,2,10))
model_fit = model.fit()

# Forecast future sales for the next 10 years
forecast = model_fit.forecast(steps=10*12)

# Plot the forecasted sales
plt.plot(sales)
plt.plot(forecast)
plt.title('Stock Forecast for the Next 10 Years')
plt.xlabel('Year')
plt.ylabel('Stocks')
plt.legend(['Forecasted Sales', 'Historical Sales'])
plt.show()
```
